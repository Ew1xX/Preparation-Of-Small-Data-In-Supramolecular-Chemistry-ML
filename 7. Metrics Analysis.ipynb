# Importing data

```python
#metrics based on WP6 dataset
WP6 = pd.read_csv("/content/drive/MyDrive/Files For Project/METRICS_WP6.csv")
WP6 = WP6.drop(["Unnamed: 0"], axis = 1)
#metrics based on WP6++ dataset
WP6_add = pd.read_csv("/content/drive/MyDrive/Files For Project/METRICS_WP6++.csv")
WP6_add  = WP6_add .drop(["Unnamed: 0"], axis = 1)
#metrics based on WP6+generated data dataset
WP6_gen = pd.read_csv("/content/drive/MyDrive/Files For Project/METRICS_WP6_gen.csv")
WP6_gen = WP6_gen.drop(["Unnamed: 0"], axis = 1)
#metrics based on WP6++ +generated data dataset
WP6_add_gen  = pd.read_csv("/content/drive/MyDrive/Files For Project/METRICS_WP6++_gen.csv")
WP6_add_gen  = WP6_add_gen .drop(["Unnamed: 0"], axis = 1)
```

# Combining all metrics

❗ Reminder:

---

* 0-2 rows - WP6

* 9-11 rows - WP6 + 50% generated data

```python
#combining all datasets
combined_df = pd.concat([WP6, WP6_add, WP6_gen, WP6_add_gen], ignore_index=True)
combined_df.to_csv("METRICS_all.csv")

combined_df
```

# Searching the top-1 metric

```python
min_value = combined_df.min().min()
min_index = combined_df.stack().idxmin()

print("Minimal value of metrics dataset:", min_value)
print("Location of min value:", min_index)
```

The best model which has this metric is kNN model and it was trained on WP6++ dataset

# Top-10 metrics (MSE)

Only MSE as it is the lowest of the 3 metrics given: MSE, MAE and RMSE.

```python
min_values = combined_df.stack().nsmallest(10)
locations = min_values.index.tolist()

for idx, value in enumerate(min_values):
    print(f"Top-{idx + 1} metric's value:", value)
    print(f"Location of the top-{idx + 1} metric:", locations[idx])
    print()
```
