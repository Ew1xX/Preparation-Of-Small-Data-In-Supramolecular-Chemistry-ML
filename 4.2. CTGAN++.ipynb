# Importing data

```python
df = pd.read_csv("/content/drive/MyDrive/Files For Project/AtomPairsData++.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df
```

```python
# Determine categorical columns
categorical_features = ["209", "193"]
```

# Functions

```python
def set_seed():
  np.random.seed(42)
  torch.manual_seed(42)
  tf.random.set_seed(42)
  random.seed(42)
```

# CTGAN

### Model №1

```python
set_seed()
ctgan = CTGAN(verbose=True)
ctgan.fit(df, categorical_features, epochs = 500)
```

```python
set_seed()
samples = ctgan.sample(100)
samples
```

Dataset value bounds: from 0 to 1, so let's remove unnecessary values

```python
samples_cleaned = samples[(samples >= 0) & (samples <= 1)].dropna()
samples_cleaned
```

### Model №2
Increasing learning rate, generator decat (optimizator's coefficient) and quantity of epochs


```python
set_seed()
ctgan2 = CTGAN(verbose = True,
               generator_lr = 0.01,
               generator_decay=0.0001)
ctgan2.fit(df, categorical_features, epochs = 1000)
```

```python
set_seed()
samples2 = ctgan2.sample(100)
samples2
```

```python
#Removing values outside bounds
samples_cleaned2 = samples2[(samples2 >= 0) & (samples2 <= 1)].dropna()
samples_cleaned2
```

```python
samples_cleaned.shape, samples_cleaned2.shape
```

The quantity and quality of values has increased!

### Model №3

```python
ctgan3 = CTGAN(verbose = True,
               generator_lr = 0.1,
               generator_decay=0.001)
ctgan3.fit(df, categorical_features, epochs = 1000)
```

```python
samples3 = ctgan3.sample(100)
samples3
```

```python
#Removing values outside bounds
samples_cleaned3 = samples3[(samples3 >= 0) & (samples3 <= 1)].dropna()
samples_cleaned3
```
